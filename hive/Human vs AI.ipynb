```python
from hive import *
from plot_utils import *
from pprint import pprint
```

```python
# New Game

state = State()
players = ['ia', 'ia']
round = 0
won = None

while won is None:
    print('Round %d' % round)
    for i, player in enumerate(state.players):
        print("Player {}".format(player.name))
        the_moves = list(available_moves(state))
        if players[i] == 'ia':
            depth = 3
            inf = 2 ** 64
            t0 = time.time()
            move, _, n = minmax(state, player, depth, -inf, inf)
            t1 = time.time()
            total = t1 - t0
            print("  ", move, "after", n)
        else:
            pprint([[i,move] for i, move in enumerate(the_moves)])
            ix = int(input())
            move = the_moves[ix]
        state.do(move)
        won = winner(state)
        plt.figure()
        fig, _ = plot_grid(state)
        plt.show()
        if won is not None:
            break
    print('---')
    round += 1
print('Winner :', won)
```
